# **Gradient Descent (Quadratic Function)**

* **Imports**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

* Define the function

```python
def f(x):
    return (x-3)**2
```

* Define its derivative

```python
def df(x):
    return 2 * (x-3)
```

* `Gradient Descent` Algorithm

```python
def gradient_descent(learning_rate, initial_x, num_iterations):
    x = initial_x
    x_values = [x]
    f_values = [f(x)]

    for i in range(num_iterations):
        gradient = df(x)
        x = x - (learning_rate * gradient)
        x_values.append(x)
        f_values.append(f(x))
        print('Iteration {:d}: x = {:.10f}, f(x) = {:.10f} '.format(i + 1, x, f(x)))

    return x, x_values, f_values
```

* Parameters

```python
learning_rate = 0.1
# Starting point
initial_x = 0 
num_iterations = 20
```

* Run `Gradient Descent`

```python
optimal_x, x_values, f_values = gradient_descent(learning_rate, initial_x, num_iterations)
```

* Printing the `Optimal x` and `Minimum value of f(x)`

```python
print('Optimal x: ', optimal_x)
print('Minimum value of f(x): ', f(optimal_x))
```

* **Plotting Results**

```python
plt.figure(figsize=(6, 4))

x_range = np.linspace(-2, 8, 100)
plt.plot(x_range, f(x_range), label="f(x) = (x - 3)^2")
plt.scatter(x_values, f_values, color='red')
plt.plot(x_values, f_values, color='red', linestyle='dashed', marker='o')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gradient Descent Optimization")
plt.legend()
plt.grid(True)
plt.show()
```
